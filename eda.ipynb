```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/processed_data/diabetes.csv")
print(df.head())
```

```python
independent = df.drop(columns = ['Outcome'])
```

```python
for col in independent.columns:
    sns.histplot(data = df, x = col, hue = 'Outcome',  multiple="stack")
    plt.title(f'Distribution of the {col}')
    plt.show()
```

```python
for col in independent.columns:
    sns.boxplot(data = df, y = col, x = 'Outcome')
    plt.title(f'Distribution of the {col}')
    plt.show()  
```

```python
corr = independent.corr()
sns.heatmap(corr, annot = True, fmt = ".2f")
plt.title("Correlation Matrix for Independent Variable")
plt.show()
```
